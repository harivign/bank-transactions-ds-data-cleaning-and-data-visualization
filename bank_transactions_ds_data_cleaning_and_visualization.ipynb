```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='transactiontype')
plt.title('Transaction Type Distribution')
plt.show()
```

```python
sns.histplot(data=df, x='transactionamount', bins=30, kde=True)
plt.title('Transaction Amount Distribution')
plt.show()
```

```python
sns.boxplot(data=df, x='transactiontype', y='transactionamount')
plt.title('Transaction Amount by Type')
plt.show()
```

```python
sns.countplot(data=df, x='channel', hue='transactiontype')
plt.title('Transaction Type by Channel')
plt.xticks(rotation=45)
plt.show()
```

```python
top_locations = df['location'].value_counts().head(10)
sns.barplot(x=top_locations.values, y=top_locations.index)
plt.title('Top 10 Locations with Most Transactions')
plt.show()
```

```python
df['transaction_hour'] = pd.to_datetime(df['transactiondate']).dt.hour

sns.countplot(data=df, x='transaction_hour', hue='transactiontype')
plt.title('Transactions by Hour')
plt.show()
```

```python
correlation = df[['transactionamount', 'customerage', 'transactionduration', 'accountbalance']].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(data=df, x='accountbalance', y='transactionamount', hue='transactiontype')
plt.title('Account Balance vs. Transaction Amount')
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(df, x='transactionamount', y='accountbalance', color='transactiontype', title='Transaction Amount vs Account Balance')
fig.show()
```
